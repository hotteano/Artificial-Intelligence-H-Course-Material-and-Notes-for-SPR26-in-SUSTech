5. Normalize a 5x5 random matrix that follows the Gaussian distribution N(4,3).(Z-Score Normalization零均值归一化)

```python
arr5 = np.random.normal(4, np.sqrt(3), (5, 5))
print("Original matrix:")
print(arr5)

normalized = (arr5 - np.mean(arr5)) / np.std(arr5)
print("\nNormalized matrix:")
print(normalized)

print(f"\nMean after normalization: {np.mean(normalized):.6f}")
print(f"Std after normalization: {np.std(normalized):.6f}")
```

6. Consider a random positive 10x2 matrix representing cartesian coordinates, convert them to polar coordinates.

```python
cartesian = np.random.rand(10, 2) * 10
print("Cartesian coordinates:")
print(cartesian)

x = cartesian[:, 0]
y = cartesian[:, 1]
polar_r = np.sqrt(x**2 + y**2)
polar_theta = np.arctan2(y, x)
polar = np.column_stack((polar_r, polar_theta))
print("\nPolar coordinates (r, theta):")
print(polar)
```

7. Consider the vector [1, 2, 3, 4, 5], how to build a new vector with 3 consecutive zeros interleaved between each value?

```python
arr7 = np.array([1, 2, 3, 4, 5])
n_zeros = 3
result = np.zeros(len(arr7) + (len(arr7) - 1) * n_zeros, dtype=int)
result[::n_zeros+1] = arr7
print(result)
```
